```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## first import data for sly4, with width= 0.01, 0.02, 0.03, 0.04MeV

```python
df1 = pd.read_csv("sly40.01", header = None, delim_whitespace=True)
df2 = pd.read_csv("sly40.02", header = None, delim_whitespace=True)
df3 = pd.read_csv("sly40.03", header = None, delim_whitespace=True)
df4 = pd.read_csv("sly40.04", header = None, delim_whitespace=True)
```

```python
x1 = df1[0]
x2 = df2[0]
x3 = df3[0]
x4 = df4[0]

y1 = df1[3]
y2 = df2[3]
y3 = df3[3]
y4 = df4[3]
```

```python
plt.plot(x1, y1, label = 'width0.01')
plt.plot(x2, y2, label = 'width0.02')
plt.legend()
```

## The width 0.02 and 0.01 shows negative strength. code crash.

```python
plt.plot(x3, y3, label = 'width0.03')
plt.plot(x4, y4, label = 'width0.04')
plt.legend()
```

## width0.03 and 0.04 look fine. 
## Below do integral of 0.03 and 0.04 curves. 

```python
Integ3 = 0.0
for i in range(1, len(x3)):
#     print(i, x3[i], y3[i], Integ3)
    Integ3 = Integ3 + y3[i]*(x3[i]-x3[i-1])
print('Integral width0.03: ', Integ3)
```

```python
Integ4 = 0.0
for i in range(1, len(x4)):
    #print(i, x4[i], y4[i], Integ4)
    Integ4 = Integ4 + y4[i]*(x4[i]-x4[i-1])
print('Integral width0.04: ', Integ4)
```
